# CSV 데이터 중복 제거 스크립트
이 노트북은 `pill_id`(첫 번째 컬럼)를 기준으로 중복된 행을 제거하여 깨끗한 CSV 파일을 생성합니다.

```python
import pandas as pd
import os

# 1. 설정
input_file = '/Users/hsj/Documents/GitHub/alyak-deep-learning-app/deep_learning_prj/pill_data.csv'  # 원본 파일명
output_file = '/Users/hsj/Documents/GitHub/alyak-deep-learning-app/deep_learning_prj/pill_data_cleaned.csv'  # 정제 후 파일명

print(f"파일 처리 시작: {input_file}")

# 2. 데이터 로드 (인코딩 예외 처리)
try:
    # 공공데이터 등에서 흔히 쓰이는 cp949 시도
    df = pd.read_csv(input_file, encoding='cp949')
    print("cp949 인코딩으로 파일을 읽었습니다.")
except UnicodeDecodeError:
    # 실패 시 utf-8 시도
    df = pd.read_csv(input_file, encoding='utf-8')
    print("utf-8 인코딩으로 파일을 읽었습니다.")
except FileNotFoundError:
    print(f"에러: {input_file} 파일이 존재하지 않습니다. 경로를 확인해주세요.")
    df = None

if df is not None:
    original_count = len(df)
    
    # 3. 중복 제거 (첫 번째 컬럼 기준)
    # subset=[df.columns[0]] 은 첫 번째 컬럼(보통 pill_id)을 기준으로 중복을 체크합니다.
    df_cleaned = df.drop_duplicates(subset=[df.columns[0]], keep='first')
    
    cleaned_count = len(df_cleaned)
    
    # 4. 결과 출력 및 저장
    print(f"\n--- 정제 결과 ---")
    print(f"원래 데이터 수: {original_count}개")
    print(f"중복 제거 후 수: {cleaned_count}개")
    print(f"삭제된 중복 행: {original_count - cleaned_count}개")
    
    # 한글 깨짐 방지를 위해 utf-8-sig 사용
    df_cleaned.to_csv(output_file, index=False, encoding='utf-8-sig')
    print(f"\n성공: 정제된 데이터가 '{output_file}'로 저장되었습니다.")
```
